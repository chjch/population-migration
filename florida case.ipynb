```python
cntin_2021_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Population Migration/countyinflow2021.csv")
```

```python
# cntin_2021_df['y1_countyfips'] = cntin_2021_df['y1_countyfips'].astype(str)
fl_cntin_2021_df = cntin_2021_df.loc[(cntin_2021_df["y1_statefips"] == 12) & (cntin_2021_df["y2_statefips"] == 12)].copy()
print(f"{len(fl_cntin_2021_df)} total records in florida.")
fl_cntin_2021_df['y1_countyfips'] = (fl_cntin_2021_df['y1_countyfips'] + 1000).astype(str).str[1:]
fl_cntin_2021_df['y2_countyfips'] = (fl_cntin_2021_df['y2_countyfips'] + 1000).astype(str).str[1:]
fl_cntin_2021_df.head()
```

```python
fl_cntin_2021_df = fl_cntin_2021_df.join(cntbnd_pnt_gdf[['lon', 'lat']], on='y1_countyfips')
fl_cntin_2021_df.rename(columns={"lon": "y1_lon", "lat": "y1_lat"}, inplace=True)
fl_cntin_2021_df = fl_cntin_2021_df.join(cntbnd_pnt_gdf[['lon', 'lat']], on='y2_countyfips')
fl_cntin_2021_df.rename(columns={"lon": "y2_lon", "lat": "y2_lat"}, inplace=True)
fl_cntin_2021_df.head()
```
